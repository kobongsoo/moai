```python
import time
from utils import WebScraping

from os import sys
sys.path.append('./mpowerai')
from pympower.classes.mshaai import MShaAI

start_time = time.time()

shaai = MShaAI()
webscraping = WebScraping()

url = "https://docs.google.com/uc?export=download&id=188PZMkk8N1Jw0frvzN60gLLJCtL1BioD"
file_path = './tmp/test1.url'
tgtPath = './tmp/test1.mpower'

text:str = ""
try:
    error = webscraping.url_download(url=url, filepath=file_path)
except Exception as e:
    print(f'url_download error=>{e}')
    error = 1002
        
if error == 0:
    try:
        shaai.extract(srcPath=file_path, tgtPath=tgtPath)
        text = webscraping.readlines_file(filepath=tgtPath, min_len=20)
        if len(text) > 4000:
            text = text[0:3999]
    except Exception as e:
        print(f'extract error=>{e}')
else:
    print(f'url_download error=>{error}')
    
# 소요된 시간을 계산합니다.
end_time = time.time()
elapsed_time = "{:.2f}".format(end_time - start_time)
print(f'time:{elapsed_time}')        
              
print(text)
```
